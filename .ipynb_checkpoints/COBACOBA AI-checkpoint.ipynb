```python
import numpy as np  
import matplotlib.pyplot as plt  
import pandas as pd  
from sklearn.metrics import classification_report

test_dataset = pd.read_csv("C:/Users/USER/Documents/Kerja/Kuliah/IF 3170 AI/TBSC2/tubes2_HeartDisease_test.csv")
train_dataset = pd.read_csv("C:/Users/USER/Documents/Kerja/Kuliah/IF 3170 AI/TBSC2/tubes2_HeartDisease_train.csv")
test_dataset.rename(columns = {'Column1':'Age',
                               'Column2':'Sex',
                               'Column3':'Chest-Pain Type',
                               'Column4':'Resting Blood Pressure',
                               'Column5':'Serum Cholestrol',
                               'Column6':'Fasting Blood Sugar',
                               'Column7':'Resting ECG',
                               'Column8':'Max Heart Rate Achieved',
                               'Column9':'Excercise Induced Angina',
                               'Column10':'ST Depression Induced',
                               'Column11':'Peak Exercise ST',
                               'Column12':'Number of Major Vessels',
                               'Column13':'Thal',
                               'Column14':'Diagnose'
                              }, inplace = True)
train_dataset.rename(columns = {'Column1':'Age',
                               'Column2':'Sex',
                               'Column3':'Chest-Pain Type',
                               'Column4':'Resting Blood Pressure',
                               'Column5':'Serum Cholestrol',
                               'Column6':'Fasting Blood Sugar',
                               'Column7':'Resting ECG',
                               'Column8':'Max Heart Rate Achieved',
                               'Column9':'Excercise Induced Angina',
                               'Column10':'ST Depression Induced',
                               'Column11':'Peak Exercise ST',
                               'Column12':'Number of Major Vessels',
                               'Column13':'Thal',
                               'Column14':'Diagnose'
                              }, inplace = True)


train_dataset
```

```python
for col in ['Sex','Chest-Pain Type','Fasting Blood Sugar','Resting ECG','Excercise Induced Angina','Peak Exercise ST','Thal']:
    train_dataset[col] = train_dataset[col].astype('category')
train_dataset.dtypes
```

```python
mod_train_dataset = train_dataset.dropna()
mod_train_dataset.isnull().sum()
mod_test_dataset = test_dataset.dropna()
mod_test_dataset.isnull().sum()
```

```python
train_data = mod_train_dataset.iloc[:, :-1].values  
train_label = mod_train_dataset.iloc[:, 13].values
```

```python
from sklearn.impute import SimpleImputer
x_train_dataset = SimpleImputer(missing_values='?', strategy='constant', fill_value = np.nan)
x_train_dataset.fit(train_data)
x_train_dataset = x_train_dataset.transform(train_data)
final_train_dataset = SimpleImputer(missing_values=np.nan, strategy='median')
final_train_dataset.fit(x_train_dataset)
final_train_dataset = final_train_dataset.transform(x_train_dataset)

x_test_dataset = SimpleImputer(missing_values='?', strategy='constant', fill_value = np.nan)
x_test_dataset.fit(mod_test_dataset)
x_test_dataset = x_test_dataset.transform(mod_test_dataset)
final_test_dataset = SimpleImputer(missing_values=np.nan, strategy='median')
final_test_dataset.fit(x_test_dataset)
final_test_dataset = final_test_dataset.transform(x_test_dataset)

```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split 
# cv

def Average(lst): 
    return sum(lst) / len(lst) 

training_data, testing_data, training_label, testing_label = train_test_split(final_train_dataset, train_label, test_size = 0.2, random_state=123)
MLP_classifier = MLPClassifier(
                               activation = 'logistic',
                               max_iter = 2000, random_state=123).fit(training_data, training_label)
MLP_prediction = MLP_classifier.predict(testing_data)
accuracy = accuracy_score(testing_label, MLP_prediction)*100
print(MLP_prediction)
print(testing_label)
print(accuracy)

#accuracy_values = []
#for i in range(1, 100):
#    #x = 1/i
#    training_data, testing_data, training_label, testing_label = train_test_split(final_train_dataset, train_label, test_size = 0.1)
#    MLP_classifier = MLPClassifier(hidden_layer_sizes = (29,29),
#                                   activation = 'logistic',
#                               max_iter = 2000).fit(training_data, training_label)
#    MLP_prediction = MLP_classifier.predict(testing_data)
#    accuracy = accuracy_score(testing_label, MLP_prediction)*100
#    accuracy_values.append(accuracy)
#    print(i)
    
#print(str(Average(accuracy_values)) + '%')    
    
```
